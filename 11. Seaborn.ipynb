# .distplot( )  VS .displot( ) :
Distplot: Gives Histogram and distribution line

Displot: Gives just bar graph but it also has 3 kinds of plot:
1. Basic Histogram
2. kind='kde' --> kernel density estimation
3. kind='ecdf' -->  empirical cumulative distribution function

```python
vis2 = sns.displot(df['InternetUsers'])
```

##### 

```python
vis2 = sns.displot(df['InternetUsers'], kind='kde')
```

###### 

```python
vis2 = sns.displot(df['InternetUsers'], kind='ecdf')
```

## Note:

2. Bi-Variate analysis:  Graph using 2 variables.
>> boxplot

3.  Multi-variate analysis: plot the graph using more than 2 variables.

## BOX PLOT
A box plot, also known as a box and whisker plot, is a graph that summarizes a set of numerical data using boxes and lines to show how the data is distributed.

```python
vis2 = sns.boxplot(data=df, x='IncomeGroup', y='BirthRate')  #Bivariate
```

### In above boxplot:
1. Line haning below box is : MINIMUM-LINE 
2. Line above box is : MAXIMUM_LINE
3. Line inside the box is : AVERAGE-LINE (Its either MEAN or MEDIAN)
4. The Diamond point is : OUTLIER == Anomaly detection
  > It is the data point which is very far from other observation.
> 
  > It will impact more on Machine learning algorithm

## Linear Model Plot
Here you get the REGRESSION LINE and the SCATTER PLOT

```python
vis3 = sns.lmplot(data= df, x='InternetUsers', y='BirthRate')
```

###### 

### If you want only Scatter-plot then type "fit_reg=False" inside bracket

```python
vis3 = sns.lmplot(data= df, x='InternetUsers', y='BirthRate', fit_reg=False)
```

###### 

### If you want only regression Line then type "scatter=False" inside bracket

```python
vis3 = sns.lmplot(data= df, x='InternetUsers', y='BirthRate', scatter=False)
```

###### 

### You can differentiate it by adding labels with "hue"...You can also give different marker styles
for label we use hue in Pandas (seaborn) & legend in NumPy

```python
vis3 = sns.lmplot(data= df, x='InternetUsers', y='BirthRate', hue="IncomeGroup", markers=["o","*","x","d"])
```

###### 

### You can also use "col" to individual compare data in lmplot

```python
vis3 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate', hue="IncomeGroup", col='IncomeGroup')
```
